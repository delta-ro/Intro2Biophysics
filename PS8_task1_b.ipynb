```python
import matplotlib
import numpy as np
import math
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
time = []
num = []
for n in tqdm( range(100,5001,200) ):
    for j in range(10):
        a = n//2
        b = n//2
        t=0

        while a!=0 and b!=0:
            p2 = a * b / (n**2+n)
            p = np.random.rand()            
            if p<p2:
                a = a + 1
                b = b - 1
            elif p2<=p<2*p2:
                a = a - 1
                b = b + 1

            t = t+1
        time.append(t)
        num.append(n)
```

```python
l = []

for j in range(0,len(num)):
    A = 0
    for i in range(1, num[j]//2+1):
        A = A + 1/(num[j]-i)
    l.append(num[j]**2 *A)
```

```python
font = {'family' : 'Segoe UI',
        'weight' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

plt.figure (figsize=(10,10))
plt.plot(num, time, '.', color = '#4cd87a', label = 'numerical simulation')
plt.plot(num, l, '-', color = 'red', label = 'th. approximation')

plt.xlabel('Total number of cells')
plt.ylabel('Time steps to reach steady state')
plt.legend( )

plt.show()
```
