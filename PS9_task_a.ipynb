```python
import matplotlib
import numpy as np
import math
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

(i)

```python
mu, sigma = 0, 0.1 # mean and standard deviation
```

```python
x = [ ]
alp = [0, 0.1, 0.5, 1]
n = 10**5

for alpha in alp:
    x1 = [1]
    for i in range(1,n):
        ksi = np.random.normal(mu, sigma, 1) #Gaussian random variable
        x1.append((1-alpha)*x1[i-1] + ksi[0])   
    x.append(x1)
```

```python
font = {'family' : 'Segoe UI',
        'weight' : 'normal',
        'size'   : 25}
matplotlib.rc('font', **font)


j=1
plt.figure (figsize=(15,10))
N, bins, patches = plt.hist(x[j], bins=10, edgecolor='white', linewidth=1, fc=(j/3, 0, 1-j/3, 0.5))

sm = 0
mx = 0
for p in patches:
    sm = sm + p.get_height()
for p in patches:
    h = p.get_height()/sm
    if h>mx: 
        mx = h+0.03
    p.set_height(h)

xmin, xmax = plt.xlim()
w = p.get_width()
plt.ylim(0,mx)
plt.xlabel('x') 
plt.ylabel('Occurences\n(fraction)')
plt.title(r'$\alpha = $ %.0f' % alp[j])

plt.xticks(np.arange(xmin+w,xmax-w,w))
plt.show()
```

```python
from scipy.stats import pearsonr
from IPython.display import display, Math

print('\033[1m' + 'Pearson correlation coefficients for:')
for j in range(4):
    corr, _ = pearsonr( x[j][:-1], x[j][1:])
    display(   Math(  r'\alpha =  {} \quad \Rightarrow \quad C = {}'.format(round(alp[j],2), round(corr,3)) )  )
```
