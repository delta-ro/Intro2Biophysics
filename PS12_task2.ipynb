```python
import matplotlib
import numpy as np
import math
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
N = 50000
R=1
frac = []
s_c = np.pi*R**2 /(2*R)**2

for n in tqdm( range(1,N+1,20) ):
    c=0
    for i in range(n):
        pebble = 2*np.random.rand(2)-1
        r = np.sqrt(pebble[0]**2 + pebble[1]**2)
        if r<=R: c = c+1
    frac.append(c/n)

print ('For N =', n)
print ('Fraction of pebbles in the circle is\t\t', c/n,sep="")
print ('Fraction of circle and rectangle spaces is\t', s_c,sep="")
er = np.abs(s_c - c/n)
p=0
while er<1:
    er = er*10
    p = p-1
    
print ('Error is\t\t\t\t\t', round(er,1),'e',p,sep="") 
```

```python
font = {'family' : 'Segoe UI',
        'weight' : 'normal',
        'size'   : 20}
matplotlib.rc('font', **font)

plt.figure (figsize=(10,10))
plt.plot(range(1,N+1,20),frac, '.', color='#4cd87a', label = 'MC algorithm')
plt.plot(range(1,N+1,20),s_c*np.ones(len(frac)), '-', color='#d26fa5', label = r'$S_{circ}/S_{sqr}$')

plt.xlabel('Number of samples') 
plt.ylabel(r'$N_{in circ}/N_{in sqr}$')
plt.legend( )
plt.show()
```
