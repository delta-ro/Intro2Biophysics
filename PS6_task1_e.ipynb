```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
```

```python
meaan = []
K = range(0,30,2)
for k in tqdm( K ):
    n = pow (2, k)
    choose = np.random.randint(0,n)
    i=[]

    ii = 0
    a = 0; b = n-1
    while a!=b:
        m = round((b+a+1)/2)
        if choose<m:
            b=m-1
        else:
            a=m
        ii+=1
    i.append (ii)
    meaan.append ( round(np.mean(i)) ) 
```

```python
font = {'family' : 'Segoe UI',
        'weight' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

plt.figure (figsize=(8,8))
plt.plot (K, K, '-', linewidth=3, color='#d26fa5', label = 'minimum entropy')
plt.plot (K, meaan, 'o', color='#4cd87a', markersize=10, label = 'simulation')
plt.xlabel(r'$\log_2 n$')
plt.ylabel('number of questions to guess')
plt.legend()
plt.show( )
```
